```python
import pandas as pd
from sklearn import model_selection,metrics,linear_model,neighbors,tree,naive_bayes
import matplotlib.pyplot as plt

iris_data = pd.read_csv('iris.csv')
iris_data.head()

iris_data['Species'].value_counts()

def clean_data(data):
    data.loc[data['Species']=='Iris-setosa','Species'] = 0
    data.loc[data['Species']=='Iris-versicolor','Species'] = 1
    data.loc[data['Species']=='Iris-virginica','Species'] = 2
clean_data(iris_data)

X = iris_data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = iris_data['Species']
X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,test_size = 0.33,random_state = 4)

# KNN

'''
model knn
'''
k_values = range(1,26)
scores = {}
scoresList = []
for k in k_values:
    knn = neighbors.KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    scores[k] = metrics.accuracy_score(y_test,y_pred)
    scoresList.append(metrics.accuracy_score(y_test,y_pred))


'''
plotting k values with the prediction scores to determine the overfitting and underfitting levels 
'''
plt.plot(k_values,scoresList)
plt.xlabel('k values')
plt.ylabel('predictions')


knn = neighbors.KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
knn_pred = knn.predict(X_test)

metrics.accuracy_score(y_test,knn_pred)

# Gaussian NB

gaussNB_model = naive_bayes.GaussianNB()
gaussNB_model.fit(X_train,y_train)

gauss_pred = gaussNB_model.predict(X_test)

df = pd.DataFrame({'Actual':y_test,'Predicted':gauss_pred})

metrics.accuracy_score(y_test,gauss_pred)

# Logistic regression

logReg_model = linear_model.LogisticRegression()
logReg_model.fit(X_train,y_train)
logReg_pred = logReg_model.predict(X_test)

metrics.accuracy_score(y_test,logReg_pred)

# Decision tree

tree_model = tree.DecisionTreeClassifier()
tree_model.fit(X_train,y_train)
tree_pred = tree_model.predict(X_test)

metrics.accuracy_score(y_test,tree_pred)

iris_data['Species'].value_counts()

def clean_data(data):
    data.loc[data['Species']=='Iris-setosa','Species'] = 0
    data.loc[data['Species']=='Iris-versicolor','Species'] = 1
    data.loc[data['Species']=='Iris-virginica','Species'] = 2
clean_data(iris_data)

X = iris_data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = iris_data['Species']
X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,test_size = 0.33,random_state = 4)

# KNN

'''
model knn
'''
k_values = range(1,26)
scores = {}
scoresList = []
for k in k_values:
    knn = neighbors.KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    scores[k] = metrics.accuracy_score(y_test,y_pred)
    scoresList.append(metrics.accuracy_score(y_test,y_pred))


'''
plotting k values with the prediction scores to determine the overfitting and underfitting levels 
'''
plt.plot(k_values,scoresList)
plt.xlabel('k values')
plt.ylabel('predictions')


knn = neighbors.KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
knn_pred = knn.predict(X_test)

metrics.accuracy_score(y_test,knn_pred)

# Gaussian NB

gaussNB_model = naive_bayes.GaussianNB()
gaussNB_model.fit(X_train,y_train)

gauss_pred = gaussNB_model.predict(X_test)

df = pd.DataFrame({'Actual':y_test,'Predicted':gauss_pred})

metrics.accuracy_score(y_test,gauss_pred)

# Logistic regression

logReg_model = linear_model.LogisticRegression()
logReg_model.fit(X_train,y_train)
logReg_pred = logReg_model.predict(X_test)

metrics.accuracy_score(y_test,logReg_pred)

# Decision tree

tree_model = tree.DecisionTreeClassifier()
tree_model.fit(X_train,y_train)
tree_pred = tree_model.predict(X_test)

metrics.accuracy_score(y_test,tree_pred)
```
